# 1991 to 2019 data

```python
#Import necessary libraries:
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

#Set the path for the Chrome WebDriver executable:
path = r"C:\Users\subha\OneDrive\Desktop\chromedriver.exe"

#Create a Chrome WebDriver instance using the specified path:
with webdriver.Chrome(path) as driver:
    driver.get("http://www.howstat.com/cricket/statistics/matches/MatchList.asp?Group=2000010120001231&v=M&Range=2000")

    #Set the target text value to search for:
    target_text = 1991
    
    #Start an infinite loop:
    while True:
        #Construct the XPath expression to find the link with the target text:
        xpath_expression = f"//a[@class='LinkOff']//nobr[text()='{target_text}']"
        
        #Find the link element using the XPath expression:
        link = driver.find_element(By.XPATH, xpath_expression)
        
        #Scroll the page to bring the link into view:
        driver.execute_script("arguments[0].scrollIntoView();", link)
        
        #Pause execution for 2 seconds:
        time.sleep(2)
        
        #Click on the link:
        link.click()
        print("Target Text:", target_text)
        
        #Find the table element on the page:
        table = driver.find_element(By.XPATH, '//table[@class="TableLined"]')
        
        #Iterate over the rows of the table and extract data into a list:
        rows = []
        for row in table.find_elements(By.TAG_NAME, 'tr'):
            row_data = [data.text for data in row.find_elements(By.TAG_NAME, 'td')]
            if row_data:
                rows.append(row_data)
        
        #Create a pandas DataFrame from the extracted data:
        df = pd.DataFrame(rows)
        with open('testcricket1991-2019.csv', 'a', newline='') as f:
            df.to_csv(f, header=False, index=False)

        print(df)
        
        #Pause execution for 2 seconds:
        time.sleep(2)
        
        #Increment the target text value by 1:
        target_text += 1
        
        #Check if the next link exists by finding elements with the updated XPath expression:
        next_link_exists = driver.find_elements(By.XPATH, xpath_expression)
        
        #Exit the loop if the next link does not exist:
        if not next_link_exists:
            break
            
    #Clean up and close the web driver:
    driver.quit()
```

```python
import csv

# List of input files
input_files = ['testcricket1991-2019.csv', 'testcricket2020.csv', 'testcricket2021.csv', 'testcricket2022.csv', 'testcricket2023.csv']

# Output file
output_file = 'test_cricket_1991_to_2023.csv'

# List to store rows from input files
rows = []

# Read data from input files and append to rows list
for file_name in input_files:
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        rows.extend(list(reader))

# Write the combined data to the output file
with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(rows)

print('CSV files combined successfully.')
```

# scroll and click visible element link to next page

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

path=r'"C:\Users\subha\OneDrive\Desktop\chromedriver.exe"'
# Set up the Selenium WebDriver
driver = webdriver.Chrome(path)

website='http://www.howstat.com/cricket/statistics/matches/MatchList.asp?Group=2000010120001231&v=M&Range=2000'


# Navigate to the web page
driver.get(website)

# Wait for the desired element to be clickable
element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[nobr='2001']")))

driver.execute_script("arguments[0].scrollIntoView();", element)

# Click on the element
element.click()

# Close the WebDriver
driver.quit()
```
